```python

import numpy as np
from io import BytesIO
from tensorflow.keras.applications.mobilenet import preprocess_input, MobileNet, decode_predictions
import requests
from tensorflow.keras.preprocessing import image

mobile = MobileNet(weights='imagenet')

image_predictions = {}

for iimage in imagee:
    try:

        response = requests.get(iimage, timeout=5)
        if response.status_code == 200:

            img = image.load_img(BytesIO(response.content), target_size=(224, 224))
            img_array = image.img_to_array(img)
            index = np.expand_dims(img_array, axis=0)
            index = preprocess_input(index)

            predictions = mobile.predict(index)

            decoded_predictions = decode_predictions(predictions, top=3)
            result = []
            labels = [desc for _, desc, _ in decoded_predictions[0]]
            
            image_predictions[iimage] = labels
            print(f"Predictions for {iimage}:")
            for label in labels:
                print(f"  {label}")
        else:
            print(f"Error image: {iimage}")
    except Exception as e:
        print(f"Error processing image {iimage}: {e}")
    
```

```python
from transformers import pipeline
classifier = pipeline('zero-shot-classification', model="facebook/bart-large-mnli")
```

```python
from collections import Counter



labels = [ "Science", "Game", "Entertainment", "Sports", "Vehicles", 
          "Travel", "Food", "Politics", "Relationships", "Health", "Career", "Esports ", "Finance","History" "Education", "Environment", "Motor Sports", "Music", "Animals",  "Fashion", "Fast Food",
            "Smartphones", "Branding ", "Leaderboard", "Art"]


tech_labels = { "Science", "Game", "Esports" ,"Motor Sports", "Smartphones", "Electric Vehicles",  "Vehicles"}
trend_labels = {"Entertainment", "Sports", "Travel", "Food", "Art",
                "Music", "History", "Politics", "Relationships", "Health",  "Finance", "Education", "Environment", "Leaderboard" ,"Animals", "Fashion", "Branding"}

trend_data = []
tech_data = []
counter = Counter()
loop = set()
for setup in data:
    if setup.get('images'):
        best_label = None
        best_score = 0.0

        for img_url in setup['images']:
            if img_url in image_predictions:
                text_description = " ".join(image_predictions[img_url])
                classification = classifier(text_description, labels)


                max_index = classification['scores'].index(max(classification['scores']))
                best_label = classification['labels'][max_index]
                best_score = classification['scores'][max_index]

                print(f"Image: {img_url}")
                print(f"Best Label: {best_label} with Score: {best_score}")


        if best_label and id(setup) not in loop:
            counter[best_label] += 1 


            loop.add(id(setup))


            filtered_setup = {k: v for k, v in setup.items() if k not in ["predicted_label", "score"]}

            # Phân loại vào nhóm
            if best_label.lower() in {label.lower() for label in tech_labels}:
                tech_data.append(filtered_setup)
            elif best_label.lower() in {label.lower() for label in trend_labels}:
                trend_data.append(filtered_setup)
```

```python
for label, count in counter.items():
    print(f"{label}: {count}")
```

```python
with open("trend.json", "w", encoding="utf-8") as trend_file:
    json.dump(trend_data, trend_file, ensure_ascii=False, indent=4)

with open("tech.json", "w", encoding="utf-8") as tech_file:
    json.dump(tech_data, tech_file, ensure_ascii=False, indent=4)
```
